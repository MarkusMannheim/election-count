```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

sns.set_style("whitegrid")

data = pd.DataFrame(columns=["electorate", "Elections ACT", "data file"])
data.electorate = ["Brindabella", "Ginninderra", "Kurrajong", "Murrumbidgee", "Yerrabi"]
data["Elections ACT"] = [55413, 56025, 50600, 53754, 53459]
data.set_index("electorate", inplace=True)

for electorate in data.index:
    temp_data = pd.read_csv(f"./data/{electorate}Total.txt")
    number_votes = len(temp_data.pindex.unique())
    data.at[electorate, "data file"] = number_votes

fig, ax = plt.subplots(figsize=(8,8))
data.plot(kind="bar", ax=ax, width=.9)
ax.tick_params(axis="x", rotation=0)
ax.xaxis.label.set_visible(False)
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter("{x:,.0f}"))
ax.tick_params(axis="both", labelsize=14)
plt.legend(fontsize=14)
plt.tight_layout()
plt.savefig("data_check.png")
```
